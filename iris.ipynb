```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
del df['Id']
```

```python
df=df.replace({'Iris-setosa':1,'Iris-versicolor':2,'Iris-virginica':3})
df['Species']=df['Species'].astype(str).astype(int)
df.head()
```

```python
sns.countplot(df['Species'])
```

```python
sns.pairplot(df)
```

```python
km = KMeans(n_clusters=3,random_state=42)
km.fit(df)
```

```python
cluster=km.labels_
cluster
```

```python
plt.scatter(df['SepalLengthCm'],df['SepalWidthCm'],c=cluster,cmap='viridis')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('SepalLengthCm vs SepalWidthCm')
plt.show()
```

```python
plt.scatter(df['PetalLengthCm'],df['PetalWidthCm'],c=cluster,cmap='viridis')
plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')
plt.title('PetalLengthCm vs PetalWidthCm')
plt.show()
```
